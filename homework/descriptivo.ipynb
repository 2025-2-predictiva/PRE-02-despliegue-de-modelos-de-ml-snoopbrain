```python
import pandas as pd
import os

# Ruta al archivo CSV
file_path = '../files/input/house_data.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Mostrar las columnas del dataset
print("Columnas del dataset house_data.csv:")
print("=" * 50)
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

print(f"\nTotal de columnas: {len(df.columns)}")
print(f"Total de filas: {len(df)}")

# Mostrar información básica del dataset
print("\nInformación del dataset:")
print("=" * 50)
df.info()
```

```python
# Mostrar las primeras 5 filas del dataset
print("Primeras 5 filas del dataset:")
print("=" * 50)
df.head()
```

```python
# Estadísticas descriptivas del dataset
print("Estadísticas descriptivas:")
print("=" * 50)
df.describe()
```

```python
# Seleccionar las columnas como features
features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'condition']

# Crear el target con la columna price
target = 'price'

# Crear el DataFrame con las features seleccionadas
X = df[features]

# Crear la serie con el target
y = df[target]

print("Features seleccionadas:")
print("=" * 30)
print(features)
print(f"\nTarget: {target}")

print(f"\nForma del dataset de features (X): {X.shape}")
print(f"Forma del target (y): {y.shape}")

print("\nPrimeras 5 filas de las features:")
print("=" * 40)
X.head()
```

```python
# Información sobre el target (price)
print("Información sobre el target (price):")
print("=" * 40)
print(f"Valor mínimo: ${y.min():,.2f}")
print(f"Valor máximo: ${y.max():,.2f}")
print(f"Valor promedio: ${y.mean():,.2f}")
print(f"Valor mediano: ${y.median():,.2f}")
print(f"Desviación estándar: ${y.std():,.2f}")

print("\nPrimeros 10 valores del target:")
print("=" * 40)
print(y.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de los plots
plt.style.use('default')
sns.set_palette("husl")

# Crear subplots para múltiples gráficos
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Análisis del Target (Price) vs Features', fontsize=16, fontweight='bold')

# 1. Distribución del precio
axes[0, 0].hist(y, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribución del Precio')
axes[0, 0].set_xlabel('Precio ($)')
axes[0, 0].set_ylabel('Frecuencia')
axes[0, 0].ticklabel_format(style='plain', axis='x')

# 2. Precio vs Área habitable (sqft_living)
axes[0, 1].scatter(X['sqft_living'], y, alpha=0.5, color='coral')
axes[0, 1].set_title('Precio vs Área Habitable')
axes[0, 1].set_xlabel('Área Habitable (sqft)')
axes[0, 1].set_ylabel('Precio ($)')
axes[0, 1].ticklabel_format(style='plain', axis='y')

# 3. Precio vs Número de habitaciones
bedroom_price = df.groupby('bedrooms')['price'].mean()
axes[1, 0].bar(bedroom_price.index, bedroom_price.values, color='lightgreen', alpha=0.8)
axes[1, 0].set_title('Precio Promedio por Número de Habitaciones')
axes[1, 0].set_xlabel('Número de Habitaciones')
axes[1, 0].set_ylabel('Precio Promedio ($)')
axes[1, 0].ticklabel_format(style='plain', axis='y')

# 4. Precio vs Condición
condition_price = df.groupby('condition')['price'].mean()
axes[1, 1].bar(condition_price.index, condition_price.values, color='gold', alpha=0.8)
axes[1, 1].set_title('Precio Promedio por Condición')
axes[1, 1].set_xlabel('Condición')
axes[1, 1].set_ylabel('Precio Promedio ($)')
axes[1, 1].ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```

```python
# Más visualizaciones del target
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Análisis Adicional del Target (Price)', fontsize=16, fontweight='bold')

# 1. Box plot del precio por número de baños
df.boxplot(column='price', by='bathrooms', ax=axes[0, 0])
axes[0, 0].set_title('Distribución del Precio por Número de Baños')
axes[0, 0].set_xlabel('Número de Baños')
axes[0, 0].set_ylabel('Precio ($)')
axes[0, 0].ticklabel_format(style='plain', axis='y')

# 2. Precio vs Área del lote
axes[0, 1].scatter(X['sqft_lot'], y, alpha=0.3, color='purple')
axes[0, 1].set_title('Precio vs Área del Lote')
axes[0, 1].set_xlabel('Área del Lote (sqft)')
axes[0, 1].set_ylabel('Precio ($)')
axes[0, 1].ticklabel_format(style='plain', axis='y')

# 3. Precio promedio por número de pisos
floors_price = df.groupby('floors')['price'].mean()
axes[1, 0].bar(floors_price.index, floors_price.values, color='orange', alpha=0.8)
axes[1, 0].set_title('Precio Promedio por Número de Pisos')
axes[1, 0].set_xlabel('Número de Pisos')
axes[1, 0].set_ylabel('Precio Promedio ($)')
axes[1, 0].ticklabel_format(style='plain', axis='y')

# 4. Precio promedio por waterfront
waterfront_price = df.groupby('waterfront')['price'].mean()
waterfront_labels = ['Sin Vista al Agua', 'Con Vista al Agua']
axes[1, 1].bar(waterfront_labels, waterfront_price.values, color=['lightcoral', 'lightblue'], alpha=0.8)
axes[1, 1].set_title('Precio Promedio por Vista al Agua')
axes[1, 1].set_ylabel('Precio Promedio ($)')
axes[1, 1].ticklabel_format(style='plain', axis='y')
plt.setp(axes[1, 1].get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Matriz de correlación entre features y target
plt.figure(figsize=(10, 8))

# Crear un DataFrame con las features y el target para calcular correlaciones
correlation_data = X.copy()
correlation_data['price'] = y

# Calcular la matriz de correlación
correlation_matrix = correlation_data.corr()

# Crear el heatmap
sns.heatmap(correlation_matrix, 
            annot=True, 
            cmap='coolwarm', 
            center=0,
            square=True,
            fmt='.3f',
            cbar_kws={'label': 'Coeficiente de Correlación'})

plt.title('Matriz de Correlación: Features vs Target (Price)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Mostrar las correlaciones con el precio ordenadas
print("Correlaciones con el precio (ordenadas por importancia):")
print("=" * 60)
price_correlations = correlation_matrix['price'].drop('price').sort_values(key=abs, ascending=False)
for feature, corr in price_correlations.items():
    print(f"{feature:15s}: {corr:6.3f}")
```
