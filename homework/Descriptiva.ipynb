```python
import pandas as pd
import os

# Ruta al archivo CSV
file_path = '../files/input/house_data.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Mostrar las columnas del dataset
print("Columnas del dataset house_data.csv:")
print("=" * 50)
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

print(f"\nTotal de columnas: {len(df.columns)}")
print(f"Total de filas: {len(df)}")

# Mostrar información básica del dataset
print("\nInformación del dataset:")
print("=" * 50)
df.info()
```

```python
# Seleccionar las columnas como features
features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'condition']

# Crear el target con la columna price
target = 'price'

# Crear el DataFrame con las features seleccionadas
X = df[features]

# Crear la serie con el target
y = df[target]

print("Features seleccionadas:")
print("=" * 30)
print(features)
print(f"\nTarget: {target}")

print(f"\nForma del dataset de features (X): {X.shape}")
print(f"Forma del target (y): {y.shape}")

print("\nPrimeras 5 filas de las features:")
print("=" * 40)
X.head()
```

```python
# Información sobre el target (price)
print("Información sobre el target (price):")
print("=" * 40)
print(f"Valor mínimo: ${y.min():,.2f}")
print(f"Valor máximo: ${y.max():,.2f}")
print(f"Valor promedio: ${y.mean():,.2f}")
print(f"Valor mediano: ${y.median():,.2f}")
print(f"Desviación estándar: ${y.std():,.2f}")

print("\nPrimeros 10 valores del target:")
print("=" * 40)
print(y.head(10))
```
